```python
import random
#int(random.random()*45)
for i in range(6):# 0~5번까지
    num=random.randrange(1,46)#복원추출:중복될수있음 <->비복원추출
    print(num,end='/')
#난수
#https://python.flowdas.com/library/random.html
#enumerate 는 자동으로 0,1,2,3,4.....숫자추출
```

```python
food='비빔밥,불고기,김치찌개,돌솥비빔밥,된장찌개,간장게장,순두부찌개,헤물파전'
foodList=food.split(',')
dayList='월,화,수,목,금'.split(',')
nameLen=len(foodList)

for cnt,day in enumerate(dayList):
    foodPrn=random.choice(foodList)
    print(cnt+1, '번', day, '==>', foodPrn)
    foodList.remove(foodPrn)
```

```python
################################################
# 절대경로
# 윈도우에는 \문자를 , 특수제어기능으로 인기하기때문
# r을 넣어서 디렉토리 구분문자라고 명확하게 설정함
################################################

file=r"C:\Users\USER\Desktop\수업\12-27\img\flower01.jpg"

from PIL import Image
img=Image.open(file)
img.size #너비,행 408 높이,열 308
print(type(img))
```

```python
####################
#이미지를 벡터화 함
#인공지능은 백터자료만 취급
#내맘대로 이미지를 변형하고자 할떄>>PIL에서 제공되지 않는 작업중
####################
import numpy as np
imgArr=np.array(img)
print(type(imgArr))
imgArr.shape#행 열 RGB

tmp=imgArr[:100,:,:]#0번행 99행까지 100개의 모든 행의자료중 모든열 모든 RGB를다 보여줘
tmp=imgArr[:,30:50,:]#모든행과 30:49까지 열 모든 RGB
#imgArr[:-100,:,:]=0
imgArr[100:,:,:]=[255,0,0]#이작업은 array구조만 가능 가로세로 값이 같아야함 
```

```python
#matplotlib 폴더의 pyplot.py를 실행하자
import matplotlib.pyplot as plt
plt.imshow(tmp)
plt.imshow(imgArr)
```
