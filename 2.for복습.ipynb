```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re

# Provided speech
speech = '''
Ladies and gentlemen, esteemed guests,
I stand before you today to shed light on a critical aspect of our health—postprandial blood glucose. In the fast-paced rhythm of our lives, it's easy to underestimate the impact of what happens in the hours following our meals.
Postprandial blood glucose, the measurement of our blood sugar levels after eating, holds the key to understanding and managing our overall well-being. As we indulge in the pleasures of our meals, our bodies engage in a complex process, breaking down carbohydrates into glucose.
The significance of this process lies not only in the immediate satisfaction of our taste buds but in its profound effects on our health. Elevated postprandial blood glucose levels have been associated with various health challenges, including diabetes and cardiovascular issues.
Awareness is the first step towards a healthier future. By recognizing the importance of monitoring postprandial blood glucose, we empower ourselves to make informed choices. It's not just a matter of medical concern; it's a proactive approach to a healthier lifestyle.
As we navigate the choices laid out before us, let us be mindful of the impact each decision has on our postprandial blood glucose levels. Balanced meals, appropriate portion sizes, and regular physical activity become not just recommendations but pillars of a robust health regimen.
In conclusion, let us embrace the responsibility we hold for our health. By understanding and managing our postprandial blood glucose, we embark on a journey towards a future where well-being is not just a hope but a tangible reality.
Thank you for your attention.
'''

# Remove non-alphabetic characters and convert to lowercase
cleaned_text = re.sub(r'[^a-zA-Z\s]', '', speech).lower()

# Generate the word cloud 워드클라우드생성
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cleaned_text)

# Plot the WordCloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import calendar

year=2023
month=12

cal= calendar.month(year, month)
print(f"{year}년 {month}월 달력:\n")
print(cal)
```

```python
import calendar
from PIL import Image, ImageDraw, ImageFont

year=2023
# print(calendar.calendar(year))
#이미지 생성
img_width, img_height= 800,600
background_color = (255, 255, 255) # 배경색
img = Image.new('RGB', (img_width, img_height), color=background_color)
draw = ImageDraw.Draw(img)

#폰트 및 위치 설정
font=ImageFont.load_default()
text_color=(0,0,0)
x, y = 10, 10 # 시작 위치
draw.text((x, y), cal, fill=text_color, font=font)
img.save(f'calendar_{year}.png') # user/usr 폴더에 기록됨
```
